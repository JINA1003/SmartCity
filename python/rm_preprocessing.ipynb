#### 공급 예비율 전처리
공급 예비율 일별 데이터 -> 월별 평균으로<br>
**출처**: https://epsis.kpx.or.kr/epsisnew/selectEkgeEpsMepChart.do?menuId=030100<br>
시점: 2004-2026

```python
import pandas as pd

df = pd.read_csv("../data/HOME_전력수급_전력수급실적.csv", encoding='cp949')

# 필요없는 컬러 제거
rate = df.drop(columns=['일', '설비용량(MW)','최소전력(MW)','최대전력기준일시','최소전력기준일시']).copy()

# 결측 처리
rate = rate.dropna(subset=['공급예비율(%)'])

rate.to_csv('../data/output/공급예비율_월별_20052026.csv', index=False)
print("저장 완료 !")

rate.head()
```
